User Defined Functions

```python
import pandas as pd
import numpy as np

df = pd.read_excel("employee_data_45_rows.xlsx")
print(df.columns)

column_data = df["Age"].tolist()

```

Mean

```python
def mean(col):
     return sum(col) / len(col)

print("Mean =", mean(column_data))
```

median

```python
def median(col):
    s = sorted(col)
    n = len(s)
    if n % 2 == 0:
        return (s[n//2 - 1] + s[n//2]) / 2
    else:
        return s[n//2]

print("Median =", median(column_data))
```

Mode

```python
def mode(col):
    freq = {}
    for i in col:
        freq[i] = freq.get(i, 0) + 1
    max_freq = max(freq.values())
    return [k for k, v in freq.items() if v == max_freq]

print("Mode =", mode(column_data))
```

Variance

```python
def variance(col):
    m = mean(col)
    total = 0
    for i in col:
        total += (i - m) ** 2
    return total / len(col)

print("Variance =", variance(column_data))
```

Standard Deviation

```python
import math

def std_dev(col):
    return math.sqrt(variance(col))

print("Standard Deviation =", std_dev(column_data))
```

Min

```python
def minimum(col):
    min_val = col[0]
    for x in col:
        if x < min_val:
            min_val = x
    return min_val

print("Minimum =", minimum(column_data))
```

Max

```python
def maximum(col):
    max_val = col[0]
    for x in col:
        if x > max_val:
            max_val = x
    return max_val

print("Maximum =", maximum(column_data))
```

Range

```python
def data_range(col):
    return maximum(col) - minimum(col)

print("Range =", data_range(column_data))
```

Histogram

```python
import matplotlib.pyplot as plt

plt.hist(column_data, bins=10)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age ")
plt.show()
```
